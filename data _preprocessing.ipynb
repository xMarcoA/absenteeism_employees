```python
import pandas as pd
```

```python
raw_data = pd.read_csv("./data/Absenteeism_data.csv")
```

```python
raw_data.info()
```

```python
pd.options.display.max_columns = None
#pd.options.display.max_rows = None
display(raw_data)
```

```python
raw_data = raw_data.drop(axis=1, columns="ID")
```

```python
dummies = pd.get_dummies(raw_data["Reason for Absence"], drop_first=True)
```

```python
# 1-14  = diseases
# 15-17 = pregnancy and giving birth
# 18-21 = poisoning and co. 
# 22-28 = light reasons
```

```python
absence_diseases = dummies.loc[:,:14].max(axis=1)
absence_pregnancy = dummies.loc[:,15:17].max(axis=1)
absence_poisoning = dummies.loc[:,18:21].max(axis=1)
absence_light_reasons = dummies.loc[:,22:].max(axis=1)
```

```python
absence_diseases.head()
```

```python
data = raw_data
```

```python
data["Absence diseases"]= absence_diseases
data["Absence pregnancy"]= absence_pregnancy
data["Absence poisoning"]= absence_poisoning
data["Absence light reasons"]= absence_light_reasons
```

```python
data['DateTime'] = pd.to_datetime(data['Date'])
data['Month'] = data['DateTime'].dt.month
data['Weekday'] = data['DateTime'].dt.weekday
data = data.drop(axis=1, columns=["Date","DateTime","Reason for Absence"])
```

```python
b = []
for value in data["Education"]:
    if value >= 2:
        b.append(1)
    else:
        b.append(0)
data["Education"]=b
```

```python
data.to_csv("./data/prep.csv")
data.head()
```
